```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("df-en-final.csv")
```

```python
count={}
```

```python
for i in range(len(df)):
    for i in df.loc[i,"P-Ingredients"].split(","):
        if i in count.keys():
            count[str(i)]+=1
        else:
            count[str(i)]=1    
            
```
