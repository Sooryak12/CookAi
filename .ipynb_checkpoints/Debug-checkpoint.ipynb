```python
user_ingredients="Chicken, Onion , Tomato , Green Chillies slit, inch Ginger , cloves Garlic , Turmeric powder (Haldi)"
```

```python
df1.reset_index(drop=True,inplace=True)
```

```python
df1.head()
df1["flag"]=0
```

```python
for b in range(5):
        flag=0
        for j in  user_ingredients.split(","):
              for i in df1.loc[b,"TranslatedIngredients"].split(","):
             #for j in  user_ingredients.split(","):
                print(i,j)
                print(fuzz.token_set_ratio(i,j))
                if fuzz.token_set_ratio(i,j)>80:
                        flag=flag+1
                        break
                
                print(flag)        
        df1.loc[b,"flag"]=flag
        print( df1.loc[b,"flag"])
```
